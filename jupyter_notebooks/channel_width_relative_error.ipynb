# Plotting the relative error between the set and measured channel width

## Definition of the Relative Error of the Channel Width
Let the set channel width be $W_s$ and the measured channel width be $W_m$, then the relative error is $$err_{rel} = \frac{W_m - W_s}{W_s} $$

The error has been computed in the spreadsheet.

```python
import os
import matplotlib.pyplot as plt
```

```python
glob_figpath = '/home/mkrause/data/artificial-muscle/channel_width_relative_err'
if not os.path.exists(glob_figpath):
    os.makedirs(glob_figpath)
```

```python
def scatter_plot(x:list, y:list, titel:str=None, xlabel:str=None, ylabel:str=None, figname:str=None) -> None:
    fig, ax = plt.subplots()
    ax.scatter(x, y)
    if title:
        ax.set_title(title)
    if xlabel:
        ax.set_xlabel(xlabel)
    if ylabel:
        ax.set_ylabel(ylabel)
    #if xlim:
    #plt.xlim(xlim)
    if figname:
        figpath = os.path.join(glob_figpath, figname)
        fig.savefig(figpath)
```

```python
set_width = [96, 192, 288, 384, 512, 608, 704, 800, 896, 992]
xlabel = 'Set width ($\mu m$)'
ylabel = 'Relative error of measured width (%)'
```
